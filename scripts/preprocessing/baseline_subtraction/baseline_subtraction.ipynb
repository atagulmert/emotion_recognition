<h1>Baseline Subtraction from ECG Signal</h1>

Subtracts the baseline from the given ecg signal.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
import peakutils
from scipy.sparse import spdiags
```

```python
df = pd.read_csv('dataset1.csv')
df_list = np.array_split(df,6)

ecg = df_list[0]['ecg']
timestamp = df_list[0]['timest']

rcParams['figure.figsize'] = 18.5,10.5

a1 = plt.scatter(timestamp,ecg,s=2,c='r')

baseline = peakutils.baseline(ecg) #Calculates the baseline

a2 = plt.scatter(timestamp,baseline,s=2,c='b')

data=[a1,a2]
data
```
