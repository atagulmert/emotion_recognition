<h1>Low Pass Filter</h1>

Applies low pass filter to a stream of data. In this case gsr data.
<br>Used formula : filteredValue = oldValue + (newValue - oldValue) / (smoothing / timeSinceLastUpdate)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
df = pd.read_csv('dataset1.csv')
df_list = np.array_split(df,6)
```

```python
gsr = df_list[0]['gsr'].tolist()
timestamp = df_list[0]['timest'].tolist()
```

```python
rcParams['figure.figsize'] = 18.5,10.5
plt.scatter(timestamp,gsr,s=2,c='r')
plt.ylabel('gsr')
plt.show()
```

```python
# values : a list of data
# smoothing: power of the filter, 1 = no change bigger the smoother
def low_pass_filter(values,smoothing):
    value = values[0]
    for index,item in enumerate(values):
        currentValue = values[index]
        value += (currentValue-value) / smoothing
        values[index] = value
gsr_2 = gsr
low_pass_filter(gsr,2)
```

<h2>Below is the data after being filtered with low pass filter using smoothing = 2</h2>

```python
plt.scatter(timestamp,gsr,s=2,c='r')
plt.ylabel('gsr')
plt.show()
```

<h2>Below is the data after being filtered with low pass filter using smoothing = 5</h2>

```python
low_pass_filter(gsr,5)
plt.scatter(timestamp,gsr_2,s=2,c='r')
plt.ylabel('gsr')
plt.show()
```
