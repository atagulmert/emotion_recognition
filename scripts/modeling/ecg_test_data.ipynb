```python
import numpy as np
import pandas as pd
import os
import os.path
```

```python
dataset = []
counter = -1

for dirpath, dirnames, filenames in os.walk(".//Downloads/Filtered_Datas/Filtered_Dataframes/"):
    for filename in [f for f in filenames if f.endswith(".csv")]:
        location = os.path.join(dirpath, filename)
        #print(filename)
        dataset.append(pd.read_csv(location))
        counter += 1
        if 'anger' in filename:
            print(counter)
            dataset[counter]['emotion'] = '1'
            dataset[counter] = dataset[counter].drop(columns=['gsr','temp','hr','spo2','timest'])
            dataset[counter].to_csv('./anger/'+filename,encoding='utf-8', index = False, float_format = "%.4f")
        if 'calmness' in filename:
            print(counter)
            dataset[counter]['emotion'] = '0'
            dataset[counter] = dataset[counter].drop(columns=['gsr','temp','hr','spo2','timest'])
            dataset[counter].to_csv('./calmness/'+filename,encoding='utf-8', index = False, float_format = "%.4f")
        if 'disgust' in filename:
            print(counter)
            dataset[counter]['emotion'] = '2'
            dataset[counter] = dataset[counter].drop(columns=['gsr','temp','hr','spo2','timest'])
            dataset[counter].to_csv('./disgust/'+filename,encoding='utf-8', index = False, float_format = "%.4f")
        if 'fear' in filename:
            print(counter)
            dataset[counter]['emotion'] = '3'
            dataset[counter] = dataset[counter].drop(columns=['gsr','temp','hr','spo2','timest'])
            dataset[counter].to_csv('./fear/'+filename,encoding='utf-8', index = False, float_format = "%.4f")
        if 'happiness' in filename:
            print(counter)
            dataset[counter]['emotion'] = '4'
            dataset[counter] = dataset[counter].drop(columns=['gsr','temp','hr','spo2','timest'])
            dataset[counter].to_csv('./happiness/'+filename,encoding='utf-8', index = False, float_format = "%.4f")
        if 'sadness' in filename:
            print(counter)
            dataset[counter]['emotion'] = '5'
            dataset[counter] = dataset[counter].drop(columns=['gsr','temp','hr','spo2','timest'])
            dataset[counter].to_csv('./sadness/'+filename,encoding='utf-8', index = False, float_format = "%.4f")
        else:
            print('Out of bound')
#dataset = pd.read_csv(".//Downloads/Filtered_Datas/Filtered_Dataframes/Participant01/anger/s01_anger_video01.csv")
```
